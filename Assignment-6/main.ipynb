### Task: Plot Mathematical Functions Based on User Input

1. **Input**: Prompt the user to enter a value for `M` (e.g., `M` is a multiplier for the functions).
2. **Generate Data**:
    - Use `np.linspace()` to generate `x` values ranging from -10 to 10.
3. **Compute Functions**:
    - Compute `y1` for the function \( y = M \c dot x^2 \).
    - Compute `y2` for the function \( y = M \c dot \sin(x) \).
4. **Plot**:

    - Use different colors and line styles for each function.
    - Add a legend to distinguish the functions.
    - Include a grid for better visualization.

```python
x=np.linspace(-10,10,1000)
```

```python
m=13
```

```python
y=m*x
plt.plot(x,y)
plt.title('y=13x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
y=m*np.sin(x)
plt.plot(x,y)
plt.title('y=13sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

### Task: Create a Dataset of Five Subjects and Scores

1. **Dataset Creation**:
    - Create a dataset containing five subjects and their respective scores.

2. **Convert to Pandas DataFrame**:
    - Use the `pandas` library to convert the dataset into a DataFrame.

3. **Plot the Scores**:
    - Use a Seaborn bar plot to visualize the scores.
    - Customize the plot with the following:
      - Different colors for each bar.
      - Annotations on top of each bar to display the score values.
      - Add a title, axis labels, and a grid for better readability.

```python
# Create a dataset of five subjects and scores
data = {'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Score': [85, 90, 78, 88, 92]}

# Convert the dataset into a Pandas DataFrame
df_scores = pd.DataFrame(data)
df_scores
```

```python

plt.figure(figsize=(8, 6))
colors = sns.color_palette("husl", len(df_scores))  
barplot = sns.barplot(x='Subject', y='Score', data=df_scores, hue='Subject', dodge=False, palette=colors)

for index, row in df_scores.iterrows():
    barplot.text(index, row['Score'] + 1, f"{row['Score']}", ha='center', va='bottom', fontsize=10)

plt.title('Scores by Subject', fontsize=16)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Task: Generate and Visualize Random Dataset

1. **Set Seed**:
    - Use your roll number as the seed for NumPy's random number generator.

2. **Generate Dataset**:
    - Create a dataset of 50 values using `np.random.randn()`.

3. **Subplot Layout**:
    - Create a 2x2 subplot layout:
        - **Line Plot**: Display the cumulative sum of the dataset.
        - **Scatter Plot**: Display the dataset with random noise.

4. **Customization**:
    - Add titles, axis labels, and grids for better visualization.

```python
np.random.seed(102367013)
```

```python
data=np.random.randn(50)
```

```python
# Line Plot: Cumulative Sum
plt.subplot(2, 2, 1)
plt.plot(np.cumsum(data), label='Cumulative Sum', color='blue')
plt.title('Cumulative Sum')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)
plt.legend()

# Scatter Plot: Dataset with Random Noise
plt.subplot(2, 2, 2)
plt.scatter(range(len(data)), data + np.random.randn(len(data)), color='orange', alpha=0.7, label='Scatter with Noise')
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### Task: Analyze and Visualize Company Sales Data

1. **Dataset**: Download the dataset from the following link:  
    [company_sales_data.csv](https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv)

2. **Visualization Tasks**:
    - **Line Plot**: Read the "Total profit" of all months and visualize it using a line plot.
    - **Multiline Plot**: Read all product sales data and visualize them using a multiline plot.
    - **Bar Chart**: Create a bar chart for all the features/attributes in the dataset.

3. **Steps**:
    - Load the dataset using `pandas`.
    - Use the `seaborn` library for visualization.
    - Customize the plots with titles, labels, legends, and grids for better readability.

```python
df = pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')
```

```python
df
```

```python
sns.lineplot(data=df, x='month_number', y='total_profit', label='Total Profit', color='red', marker='o')
plt.title('Monthly Sales Analysis')
plt.xlabel('Month Number')
plt.ylabel('Total Profit/Units')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
# Assuming you want to plot all columns except 'month_number'
columns_to_plot = [col for col in df.columns if col != 'month_number']

plt.figure(figsize=(10, 6))
for col in columns_to_plot:
    sns.lineplot(data=df, x='month_number', y=col, marker='o', label=col)

plt.title('Monthly Sales Analysis')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
df_melted = df.melt(id_vars='month_number', var_name='Feature', value_name='Value')

plt.figure(figsize=(12, 8))
sns.barplot(data=df_melted, x='Feature', y='Value', hue='month_number', palette='viridis')
plt.title('Bar Chart for All Features/Attributes', fontsize=16)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(title='Month Number', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
