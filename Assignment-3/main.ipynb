- Sort the DataFrame by the Salary column in descending order. 

```python
df.sort_values(by='Salary', ascending=False,inplace=True)
df
```

- Add a new column that categorizes employees based on their performance rating: 
    -   Excellent for ratings >= 4.5 
    -   Good for ratings >= 4.0 but < 4.5 
    -   Average for ratings < 4.0  

```python
def performance_category(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'
```

```python
df['Performance'] = df['Ratings'].apply(performance_category)
df[['Name', 'Ratings', 'Performance']]
```

- Identify missing values in the DataFrame.


- Rename the Employee_ID column to ID.


- Modify the dataset by adding a new column, Tax, which deducts 10% of the  Salary.
